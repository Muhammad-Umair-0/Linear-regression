###  feature selection through      Chi Square (categorical Attributes)

```python
df = pd.read_csv('Loan Prediction Dataset.csv')
df = df[['Gender','Married','Dependents','Education', 'Self_Employed','Credit_History','Property_Area','Loan_Status']]
df.head()
```

```python
for col in df.columns:
    df[col] = df[col].fillna(df[col].mode()[0])

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
from sklearn.feature_selection import chi2
X = df.drop(columns=['Loan_Status'], axis=1)
y= df['Loan_Status']

```

```python
chi_score = chi2(X,y)
```

```python
chi_score
```

```python
# higher chi value ,higher the importance 
chi_values = pd.Series(chi_score[0], index=X.columns)
chi_values.sort_values(ascending = False, inplace=True)
chi_values.plot.bar()
```

```python
# if p value >0.5 ,lower the importance 
chi_values = pd.Series(chi_score[1], index=X.columns)
chi_values.sort_values(ascending = False, inplace=True)
chi_values.plot.bar()
```

## Cross Validation Techniques 

#### K-Fold Cross Validation 

```python
df= df.drop(['instant','dteday','casual','registered'],axis=1)
```

```python
df.head()
```

```python
X = df.drop('cnt',axis=1)
y = df['cnt']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
```

```python
cv = KFold(n_splits=5, random_state=42,shuffle=True)
model =RandomForestRegressor()
scores = cross_val_score(model,X,y, cv=cv, n_jobs = -1)
print(f"Error MEan {np.mean(scores)} Error Std {np.std(scores)}")
```

## Repeated stratified KFold Cross validation 

```python
from sklearn.model_selection import RepeatedStratifiedKFold
cv = RepeatedStratifiedKFold(n_splits=5,n_repeats=3 , random_state=42)
model =RandomForestRegressor()
scores = cross_val_score(model,X,y, cv=cv, n_jobs = -1)
print(f"Error MEan {np.mean(scores)} Error Std {np.std(scores)}")
```
